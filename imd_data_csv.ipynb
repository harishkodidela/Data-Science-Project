```python
import imdlib as imd

path = "/content/drive/MyDrive/Colab Notebooks"

start_yr = 1972
end_yr = 2022
var1 = 'rain' # rain data variable
var2 = 'tmin' # minimum temperature variable
var3 = 'tmax' # maximum temperature variable

# getting data from imdlib package
imd.get_data(var1, start_yr, end_yr, fn_format='yearwise', file_dir=path)
imd.get_data(var2, start_yr, end_yr, fn_format='yearwise', file_dir=path)
imd.get_data(var3, start_yr, end_yr, fn_format='yearwise', file_dir=path)
data1 = imd.open_data(var1, start_yr, end_yr,'yearwise', path)
data2 = imd.open_data(var2, start_yr, end_yr,'yearwise', path)
data3 = imd.open_data(var3, start_yr, end_yr,'yearwise', path)
ds1 = data1.get_xarray()
ds2 = data2.get_xarray()
ds3 = data3.get_xarray()
print(ds1,ds2,ds3)
```

```python
lat = 12.9716 #lattitude of point
lon = 77.5946 #longitude of point

# data coverting to csv file
data1.to_csv('data1.csv', lat, lon, path)
data2.to_csv('data2.csv', lat, lon, path)
data3.to_csv('data3.csv', lat, lon, path)
```
